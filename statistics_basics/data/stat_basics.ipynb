# Задание 1. Загрузка данных

```python
colnames = ['surgery', 'Age', 'Hospital Number', 'rectal temperature','pulse', 'respiratory rate',\
'temperature of extremities', 'peripheral pulse','mucous membranes','capillary refill time',\
'pain', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux',\
'nasogastric reflux PH', 'rectal examination', 'abdomen', 'packed cell volume', 'total protein',\
'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome', 'surgical lesion',\
'type of lesion1', 'type of lesion2', 'type of lesion3', 'cp_data']

df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv', names = colnames, header=None, na_values='?')
```

# Задание 2. Первичное изучение данных

```python
df.info() #Смотрим кол-во пропусков в данных и типы данных
```

```python
df.describe() #расчитываем базовые статистики
```

```python
#предполагаем опечатку в Age и заменяем 9 на 2
df['Age'].replace({9:2}, inplace = True)
```

```python
#Количественные pulse и respiratory rate имеют не нормальное распределение - видим по значению std. 
#Но из-за специфики датасета завышенные и заниженные значения могут встречаться. 
#Например, если мы уберем максимумы, то std изменится незначительно. Поэтому работаем со всеми данными.

print(df[(df['pulse'] > 30) & (df['pulse'] < 184)]['pulse'].std())
print(df[(df['respiratory rate'] > 8) & (df['respiratory rate'] < 96)]['respiratory rate'].std())
```

# Задание 3. Работа с пропусками

```python
#Посчитаем кол-во пропусков
df.isna().sum()
```

```python
#Посчитаем процент пропусков
(df.isna().mean() * 100).round(2)
```

```python
#Определяем количественные непрерывные атрибуты
for col in df.columns.values:
    if (len(df[col].value_counts())> 5) and (df[col].isnull().sum() > 0):
        print(col)
```

```python
#Количественные заменяем на их медианные значения 
df['rectal temperature'].fillna(df['rectal temperature'].median(), inplace = True)
df['pulse'].fillna(df.groupby('pain')['pulse'].transform('median'), inplace = True) #чем сильнее боль, тем выше пульс
df['pulse'].fillna(df['pulse'].median(), inplace = True) #в 8 строках нет атрибута pain , поэтому просто медиану берем для них
df['respiratory rate'].fillna(df['respiratory rate'].median(), inplace = True) #большой разброс значений, бесполезный атрибут
```

```python
#Для категориальных переменных используем моду
df['surgery'].fillna(df['surgery'].mode()[0], inplace = True) #всего 1 значение пропущено
df['temperature of extremities'].fillna(df['temperature of extremities'].mode()[0], inplace = True)
df['pain'].fillna(df['pain'].mode()[0], inplace = True)
df['outcome'].fillna(df['outcome'].mode()[0], inplace = True)
```

```python
#Проверяем пропуски
df.isna().sum()
```
