````python
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

if len(phrase_1) > len(phrase_2):
    print('Фраза 1 длиннее фразы 2')
elif len(phrase_2) > len(phrase_1):
    print('Фраза 2 длиннее фразы 1')
else:
    print('Фразы равной длины')
````

## Задание 2

Дана переменная, в которой хранится число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
#Год является високосным, если его номер кратен 4, но не кратен 100, или если он кратен 400.
y = 2020
if y % 4 == 0 and y % 100 != 0 or y % 400 == 0: 
    print('Високосный год')
else:
    print('Обычный год')
````

````python
#определение знака зодиака по дню и месяцу рождения
d_of_birth, m_of_birth = int(input('Введите день: ')), str(input('Введите месяц, например, Март: '))

if 1 <= d_of_birth <= 31 and m_of_birth in \
    ('Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь', 'Январь', 'Февраль'):
    if m_of_birth == 'Март' and 21 <= d_of_birth <= 31 or m_of_birth == 'Апрель' and 1 <= d_of_birth <= 19:
        print('Ваш знак зодиака: Овен')
    elif m_of_birth == 'Апрель' and 20 <= d_of_birth <= 30 or m_of_birth == 'Май' and 1 <= d_of_birth <= 20:
        print('Ваш знак зодиака: Телец')
    elif m_of_birth == 'Май' and 21 <= d_of_birth <= 31 or m_of_birth == 'Июнь' and 1 <= d_of_birth <= 21:
        print('Ваш знак зодиака: Близнецы')
    elif m_of_birth == 'Июнь' and 22 <= d_of_birth <= 30 or m_of_birth == 'Июль' and 1 <= d_of_birth <= 22:
        print('Ваш знак зодиака: Рак')
    elif m_of_birth == 'Июль' and 23 <= d_of_birth <= 31 or m_of_birth == 'Август' and 1 <= d_of_birth <= 22:
        print('Ваш знак зодиака: Лев')
    elif m_of_birth == 'Август' and 23 <= d_of_birth <= 31 or m_of_birth == 'Сентябрь' and 1 <= d_of_birth <= 22:
        print('Ваш знак зодиака: Дева')    
    elif m_of_birth == 'Сентябрь' and 23 <= d_of_birth <= 30 or m_of_birth == 'Октябрь' and 1 <= d_of_birth <= 23:
        print('Ваш знак зодиака: Весы')     
    elif m_of_birth == 'Октябрь' and 24 <= d_of_birth <= 31 or m_of_birth == 'Ноябрь' and 1 <= d_of_birth <= 22:
        print('Ваш знак зодиака: Скорпион')    
    elif m_of_birth == 'Ноябрь' and 23 <= d_of_birth <= 30 or m_of_birth == 'Декабрь' and 1 <= d_of_birth <= 21:
        print('Ваш знак зодиака: Стрелец')    
    elif m_of_birth == 'Декабрь' and 22 <= d_of_birth <= 31 or m_of_birth == 'Январь' and 1 <= d_of_birth <= 20:
        print('Ваш знак зодиака: Козерог')      
    elif m_of_birth == 'Январь' and 21 <= d_of_birth <= 31 or m_of_birth == 'Февраль' and 1 <= d_of_birth <= 18:
        print('Ваш знак зодиака: Водолей')      
    elif m_of_birth == 'Февраль' and 19 <= d_of_birth <= 28 or m_of_birth == 'Март' and 1 <= d_of_birth <= 20:
        print('Ваш знак зодиака: Рыбы')
else:
    print('Неправильный ввод')
````

Используйте следующие правила:
- если каждое из трех измерений менее или равно 15 сантиметрам, то выведите на экран "Коробка №1";
- если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
- если длина товара больше 2 метров, то выводите "Упаковка для лыж";

Пример работы программы:

```
width = 15

```
Результат:

```
width = 45

```

`Ищите упаковку для лыж` 

````python
width = 15
length = 201
height = 15

if width <= 15 and length <= 15 and height <= 15:
    print('Коробка №1')
else:
    if length > 200:
        print('Упаковка для лыж')
    elif 15 < width < 50 or 15 < length < 50 or 15 < height < 50:
        print('Коробка №2')
    else:
        print('Стандартная коробка №3')
````

## Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Несчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
num = 123321
a = num // 100000
b = (num // 10000) % 10
c = (num // 1000) % 10
d = (num // 100) % 10
e = (num // 10) % 10
f = num % 10

if a + b + c == d + e + f:
    print('Счастливый билет')
else:
    print('Несчастливый билет')
````

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
fig = input('Введите тип фигуры: ')

if fig in ('Круг', 'Треугольник', 'Прямоугольник'):
    if fig == 'Круг':
        rad = int(input('Введите радиус круга: '))
        print('Площадь круга: ', round(3.14 * rad ** 2, 2)) # S = пR**2
    elif fig == 'Треугольник':
        a, b, c = {
            int(input('Введите длину стороны A')), 
            int(input('Введите длину стороны B')), 
            int(input('Введите длину стороны C'))
        }
        p = (a + b + c) / 2
        print('Площадь треугольника: ', round((p * (p-a) * (p-b) * (p-c)) ** 0.5, 2)) #Формула Герона
    elif fig == 'Прямоугольник':
        d, e = int(input('Введите длину стороны 1')), int(input('Введите длину стороны 2'))
        print('Площадь прямоугольника: ', d * e)
else:
    print('Ошибка ввода. Укажите правильный тип фигуры')
````
